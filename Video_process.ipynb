```python
import cv2
import os

# 비디오 파일 경로
video_path = './example.mp4'  # 비디오 파일 경로를 여기에 입력하세요.

# 비디오 파일을 엽니다.
video = cv2.VideoCapture(video_path)

# 비디오가 올바르게 열렸는지 확인
if not video.isOpened():
    print("Error: Could not open video.")
    exit()

# 비디오의 FPS (초당 프레임 수)를 얻습니다.
fps = video.get(cv2.CAP_PROP_FPS)

# 저장할 이미지의 디렉토리를 만듭니다. (동일한 디렉토리에 'frames' 폴더)
save_dir = './example'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 프레임을 하나씩 읽어서 초 단위로 이미지를 저장합니다.
count = 0
while True:
    ret, frame = video.read()
    if not ret:
        break

    # FPS를 고려하여 초 단위로 프레임을 저장합니다.
    if int(video.get(cv2.CAP_PROP_POS_FRAMES)) % int(fps/2) == 0:
        cv2.imwrite(os.path.join(save_dir, f"frame{count}.jpg"), frame)
        print(f"Saved frame{count}.jpg")
        count += 1

video.release()
```
